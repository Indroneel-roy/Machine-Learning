```python
import numpy as np

# Sample data
X = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 8])

# Parameters
w = 0
b = 0
alpha = 0.01
epochs = 1000
n = len(X)

# Gradient Descent
for i in range(epochs):
    y_pred = w*X + b
    dw = (2/n) * np.sum((y_pred - y) * X) # Gradient w.r.t. weight
    db = (2/n) * np.sum(y_pred - y) # Gradient w.r.t. bias
    w -= alpha * dw # Update weight
    b -= alpha * db # Update bias

print("Weight:", w)
print("Bias:", b)
```
